```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
pd.set_option("display.max_columns", None)

df = pd.read_csv("../data/clean_churn.csv")
df.shape
```

```python
churn_rate = df["Churn"].mean()
churn_rate
```

The overall churn rate is approximately 26%, including a moderately imbalanced classification problem.

```python
plt.figure(figsize=(6,4))
sns.barplot(
    data=df,
    x="Contract",
    y="Churn",        
    estimator=np.mean
)
plt.title("Churn Rate by Constract Type")
plt.ylabel("Churn Rate")

plt.savefig("../figures/churn_by_constract.png", bbox_inches="tight")
plt.show()
```

Customers on month-to-month contracts exhibit significantly higher hurn rates compared to customers on one and two year contracts.

```python
plt.figure(figsize=(6,4))
sns.barplot(
    data=df,
    x="tenure_group",
    y="Churn",
    estimator=np.mean
)
plt.title("Churn Rate by Tenure Group")
plt.ylabel("Churn Rate")
plt.xlabel("Tenure (months)")

plt.savefig("../figures/churn_by_tenure_group.png", bbox_inches="tight")
plt.show()
```

Churn is highest among customers with tenure with below 12 months and decreases steadily with longer tenure.

```python
plt.figure(figsize=(6,4))
sns.barplot(
    data=df,
    x="Churn",
    y="MonthlyCharges"
)
plt.title("Monthly Charges by Churn Status")
plt.xlabel("Churn (0 = No, 1 = Yes)")

plt.savefig("../figures/monthly_charges_by_churn_status.png", bbox_inches="tight")
plt.show()
```

Customers who churn tend to have higher monthly charges, suggesting price sensitivity.

```python
plt.figure(figsize=(6, 4))
sns.barplot(
    data=df,
    x="PaymentMethod",
    y="Churn",
    estimator=np.mean
)
plt.title("Churn Rate by Payment Method")
plt.ylabel("Churn Rate")
plt.xticks(rotation=30)
plt.savefig("../figures/churn_by_payment_methods.png", bbox_inches="tight")
plt.show()
```

Electronic check users have noticeably higher churn rates compared to customers using automatic payment methods.

```python
plt.figure(figsize=(6, 4))
sns.barplot(
    data=df,
    x="InternetService",
    y="Churn",
    estimator=np.mean
)
plt.title("Churn Rate by Internet Service Type")
plt.ylabel("Churn Rate")

plt.savefig("../figures/churn_by_internet_service_type.png", bbox_inches="tight")
plt.show()
```

Fiber optic customers show higher churn compared to DSL customers, potentially due to pricing or competition.

```python
num_cols = ["tenure", "MonthlyCharges", "TotalCharges", "Churn"]

plt.figure(figsize=(6, 4))
sns.heatmap(
    df[num_cols].corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)
plt.title("Correlation Matrix")

plt.savefig("../figures/correlation_matrix.png", bbox_inches="tight")
plt.show()
```

Tenure and TotalChareges are strongly correlated, while MonthlyCharges shows a weaker relationship with tenure.

**Key EDA Findings**
- Month-to-month contracts are the strongest driver of churn
- New customers (tenure < 12 months) are at highest risk
- Higher monthly charges are associated with increased churn
- Automatic payment methods are linked to lower churn
